## 05 - Faça um programa em Python ou Java que permita converter uma imagem colorida em tons de cinza e a mesma imagem colorida em preto e branco.

```python
# Codigo da Atividade 05

import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files

# faz upload das imagem
uploaded = files.upload()

# lê a imagem carregada
file_name = list(uploaded.keys())[0] # pega o nomee do arquivo carregado
originalImage = cv2.imread(file_name)

#converte a imagem para tons de cinza
grayImage = cv2.cvtColor(originalImage, cv2.COLOR_BGR2GRAY)

# Aplica um limiar para criar uma imagem em preto e branco
(threash, blackAndWhiteImage) = cv2.threshold(grayImage, 127, 255, cv2.THRESH_BINARY)

# Exibe as imagens usando cv2_imshow(especifico para colab)
print("\n")

#imagem cinza
grayImage = np.expand_dims(grayImage, axis=-1)
grayImage = np.repeat(grayImage, 3, axis=-1)

#imagem preta
blackAndWhiteImage = np.expand_dims(blackAndWhiteImage, axis=-1)
blackAndWhiteImage = np.repeat(blackAndWhiteImage, 3, axis=-1)

#print(blackAndWhiteImage.shape) #ver tamanho da imagem

r = np.concatenate((originalImage, grayImage, blackAndWhiteImage), axis=1)
cv2_imshow(r)
```
