Atividade que o professor passou (2).

1) Defina uma variável raio com valor float.

2)  calcule a área do circulo usando a formula A = PI * Raio².

3) Imprima o resultado com a mensagem: "A área do circulo é: "X" area.


```python
import math

raio = 2.5
#PI = math.pi

area = math.pi * math.pow(raio, 2)

print("A área do circulo é: ", area, ".")
```

```python
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow # para exibir imagens no calab
import matplotlib.pyplot as plt

# faz upload das imagem
uploaded = files.upload()

# lê a imagem carregada
file_name = list(uploaded.keys())[0] # pega o nomee do arquivo carregado
originalImage = cv2.imread(file_name)

#converte a imagem para tons de cinza
grayImage = cv2.cvtColor(originalImage, cv2.COLOR_BGR2GRAY)

# Aplica um limiar para criar uma imagem em preto e branco
(threash, blackAndWhiteImage) = cv2.threshold(grayImage, 127, 255, cv2.THRESH_BINARY)

# Exibe as imagens usando cv2_imshow(especifico para colab)
print("\n")
print("Imagem original:")
cv2_imshow(originalImage)

print("\n")
print("Imagem tons de cinza:")
cv2_imshow(grayImage)

print("\n")
print("Imagem tons de preto e branco:")
cv2_imshow(blackAndWhiteImage)
```
