## Exercicio de Matriz de Gaussian

O exercício mostra a matriz de gaussian, na forma original sendo composta por números e ela com cores.

```python
#Código E08-D
import numpy as np
import matplotlib.pyplot as plt

size = 5
x = np.linspace(-2, 2, size)
y = np.linspace(-2, 2, size)
x, y = np.meshgrid(x, y)

sigma = 1.0
gaussian = np.exp(-(x**2 + y**2) / (2 * sigma**2))
gaussian = gaussian / np.sum(gaussian)

print("Matriz Gaussiana de 5x5 Original")
print(gaussian)
print("")

#--------------------------------------------------------

print("Matriz Gaussiana de Multiplicada por 100")
gaussian = gaussian*100
print(gaussian)
print("")

#--------------------------------------------------------

valor = 0
for i in range(len(gaussian)):
  for j in range(len(gaussian[i])):
    gaussian[i][j] += valor
    gaussian[i][j] = int(gaussian[i][j])

print("Mtriz Gaussiana 5x5 Melhorada")
for linha in gaussian:
  print(linha)
print("")

#----------------------------------------------------------

print("Matriz Gaussiana por Cores")
plt.imshow(gaussian, cmap="viridis", extent=[-2, 2, -2, 2])
plt.colorbar()
plt.title("Matriz Gaussiana de 5x5")
plt.show()
```

```python
# Código E08-E
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files

def create_gaussian_kernel(size, sigma):
  kernel = np.zeros((size,size), dtype=np.float32)
  center = size // 2

  for i in range(size):
    for j in range(size):
      x = i - center
      y = j - center
      kernel[i, j] = np.exp(-(x**2 + y**2) / (2*sigma**2)) / (2*np.pi*sigma**2)

  kernel /= np.sum(kernel)
  return kernel

uploaded = files.upload()
filename = next(iter(uploaded))
img = cv2.imread(filename)

kernel_size = 15
sigma = 3.0
gaussian_kernel = create_gaussian_kernel(kernel_size, sigma)

filtered_img = cv2.filter2D(img, -1, gaussian_kernel)

print("Imagens")
imgs_concat = np.concatenate((img, filtered_img), axis=1)
cv2_imshow(imgs_concat)
print("")

print("Matriz de Gaussian:")
print(gaussian_kernel)
```
